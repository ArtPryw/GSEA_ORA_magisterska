```python

#DataFrame
bos_12 = pd.read_csv('dane/bronze/1_2/res-ARS-UCD1.2.109.tsv', sep='\t')
# Przygotowanie danych dla GSEA
gene_list = bos_12[['gene_name', 'log2FoldChange']].dropna()
gene_list = gene_list.sort_values(by='log2FoldChange', ascending=False) # Sortowanie genÃ³w

# Zapisanie do pliku .rnk
rnk_path = 'gene_list.rnk'
gene_list.to_csv(rnk_path, sep='\t', header=False, index=False)

# Przeprowadzenie analizy GSEA
gsea_results = gs.prerank(rnk=rnk_path,
                          gene_sets='KEGG_2019_Human',
                          min_size=2,
                          max_size=500,
                          threads=12,
                          permutation_num=2000, # liczba permutacji
                          outdir='gseapy_gsea', # folder wyjÅ›ciowy
                          seed=6)

gsea_results.res2d.to_csv('output/gsea/gsea_results.csv') 

```
