```python
bos_12 = pd.read_csv('dane/bronze/1_2/res-ARS-UCD1.2.109.tsv', sep='\t')

selected_genes = bos_12[bos_12['padj'] < 0.05]['gene_name'].dropna().tolist()

# ORA przy użyciu gseapy
enr = gp.enrichr(gene_list=selected_genes, 
                 gene_sets='KEGG_2019_Human', 
                 outdir='ora_results')

enr.results.sort_values("Combined Score")
```
